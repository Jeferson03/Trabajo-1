#### Cree una tabla **Factura** con atributos: código de la factura (clave primaria), fecha,codigo cliente y código del vendedor. Y una tabla **Detalle** con atributos: código del detalle, código de la factura (clave foránea hacia factura), código del producto, número de unidades y precio unitario.

```
DROP TABLE detalle; -- Si es la primera vez que corre el codigo esta linea suelta error pq no existe el if exists en el 21c 
DROP TABLE factura; -- Si es la primera vez que corre el codigo esta linea suelta error pq no existe el if exists en el 21c 
CREATE TABLE factura (
    codigo_factura NUMBER(10) PRIMARY KEY,
    fecha DATE NOT NULL,
    codigo_cliente NUMBER(8) NOT NULL,
    codigo_vendedor NUMBER(8) NOT NULL
);


CREATE TABLE detalle (
    codigo_detalle NUMBER(10) PRIMARY KEY,
    codigo_factura NUMBER(10) NOT NULL,
    codigo_producto NUMBER(8) NOT NULL,
    numero_unidades NUMBER(6) NOT NULL,
    precio_unitario NUMBER(8) NOT NULL,
    CONSTRAINT codigo_factura_fk FOREIGN KEY (codigo_factura) REFERENCES factura(codigo_factura)
)

```

#### Ahora creemos un procedimiento que se encargara de solucionar el item a) del punto 1 del trabajo

```
CREATE OR REPLACE PROCEDURE only_cursors
AS 
    CURSOR c_facturas IS SELECT * FROM factura; 
    CURSOR c_detalles(p_codigo_fact factura.codigo_factura%TYPE) IS SELECT * FROM detalle WHERE codigo_factura = p_codigo_fact;

    v_factura factura%ROWTYPE;
    v_detalle detalle%ROWTYPE;
    v_valor_total NUMBER(15) := 0;
    v_tiempo_inicio NUMBER;
    v_tiempo_fin NUMBER;

BEGIN 

    v_tiempo_inicio := DBMS_UTILITY.GET_TIME;
    
    FOR v_factura IN c_facturas LOOP

       

        FOR v_detalle IN c_detalles(v_factura.codigo_factura) LOOP
            
            v_valor_total := v_valor_total + v_detalle.numero_unidades * v_detalle.precio_unitario;

        END LOOP;
    
        DBMS_OUTPUT.PUT_LINE('  INFORMACION DE LA FACTURA ' || ' ' ||
            'codigo de la factura: ' || v_factura.codigo_factura || ' ' ||
            'fecha: ' || v_factura.fecha || ' ' ||
            'codigo del cliente: ' || v_factura.codigo_cliente || ' ' ||
            'codigo del vendedor: ' || v_factura.codigo_vendedor || ' ' ||
            'valor total: ' || v_valor_total 
        );

    END LOOP;

    v_tiempo_fin := DBMS_UTILITY.GET_TIME;
    DBMS_OUTPUT.PUT_LINE('TIEMPO EMPLEADO CONSULTA: ' || (v_tiempo_fin - v_tiempo_inicio)/100 || ' segundos');

    EXCEPTION
        WHEN others THEN
            DBMS_OUTPUT.PUT_LINE('Error: ' || SQLERRM);


END;
```

#### Ahora creemos un procedimiento que se encargara de las inserciones 

```
CREATE OR REPLACE PROCEDURE rellenado_tablas(p_cantidad_facturas IN NUMBER, p_cantidad_detalles IN NUMBER)
AS
    v_codigo_factura NUMBER := 1000;
    v_codigo_detalle NUMBER := 1;
BEGIN

    -- INSERTAR n FACTURAS
    FOR i IN 1..p_cantidad_facturas+1 LOOP
        
        INSERT INTO factura (
            codigo_factura,
            fecha,
            codigo_cliente,
            codigo_vendedor
        )
        VALUES (
            v_codigo_factura,
            SYSDATE - MOD(i,365),
            TRUNC(DBMS_RANDOM.VALUE(1000,9999)),
            TRUNC(DBMS_RANDOM.VALUE(200,500))
        );

        -- INSERTAR m DETALLES POR FACTURA
        FOR j IN 1..p_cantidad_detalles+1 LOOP
            
            INSERT INTO detalle (
                codigo_detalle,
                codigo_factura,
                codigo_producto,
                numero_unidades,
                precio_unitario
            )
            VALUES (
                v_codigo_detalle,
                v_codigo_factura,
                TRUNC(DBMS_RANDOM.VALUE(1,500)),
                TRUNC(DBMS_RANDOM.VALUE(1,10)),
                TRUNC(DBMS_RANDOM.VALUE(5000,100000))
            );

            v_codigo_detalle := v_codigo_detalle + 1;

        END LOOP;

        v_codigo_factura := v_codigo_factura + 1;

    END LOOP;

    COMMIT;

    DBMS_OUTPUT.PUT_LINE('Se insertaron '|| p_cantidad_facturas  ||' facturas y '|| (p_cantidad_facturas*p_cantidad_detalles) ||' detalles.');

END;
```

## Punto B

### Funcion Bulk Collect

```
CREATE OR REPLACE PROCEDURE bulk_collect_facturas(
    p_limite     IN PLS_INTEGER DEFAULT 5
)
AS

    -- Cursor para traer todas las facturas
    CURSOR c_facturas IS
        SELECT *
        FROM factura;

    -- Cursor para traer los detalles de una factura
    CURSOR c_detalles(
        p_codigo_fact factura.codigo_factura%TYPE
    ) IS
        SELECT *
        FROM detalle
        WHERE codigo_factura = p_codigo_fact;

    -- Tipo arreglo para almacenar facturas
    TYPE t_facturas IS TABLE OF factura%ROWTYPE;

    -- Variable arreglo
    v_facturas t_facturas;

    -- Variables auxiliares
    v_valor_total NUMBER(15,2);

BEGIN

    OPEN c_facturas;

    LOOP

        -- Cargar facturas por lotes
        FETCH c_facturas
        BULK COLLECT INTO v_facturas
        LIMIT p_limite;

        -- Salir cuando no haya más datos
        EXIT WHEN v_facturas.COUNT = 0;

        -- Recorrer el arreglo
        FOR i IN 1 .. v_facturas.COUNT LOOP

            v_valor_total := 0;

            -- Buscar detalles de la factura actual
            FOR v_detalle IN c_detalles(
                v_facturas(i).codigo_factura
            ) LOOP

                v_valor_total := v_valor_total +
                                  (
                                    v_detalle.numero_unidades *
                                    v_detalle.precio_unitario
                                  );

            END LOOP;

            -- Imprimir solo si el parámetro es TRUE
        

            DBMS_OUTPUT.PUT_LINE(
                ' INFORMACION DE LA FACTURA '
            );

            DBMS_OUTPUT.PUT_LINE(
                'codigo de la factura: ' ||
                v_facturas(i).codigo_factura
            );

            DBMS_OUTPUT.PUT_LINE(
                'fecha: ' ||
                v_facturas(i).fecha
            );

            DBMS_OUTPUT.PUT_LINE(
                'codigo del cliente: ' ||
                v_facturas(i).codigo_cliente
            );

            DBMS_OUTPUT.PUT_LINE(
                'codigo del vendedor: ' ||
                v_facturas(i).codigo_vendedor
            );

            DBMS_OUTPUT.PUT_LINE(
                'valor total: ' ||
                v_valor_total
            );

            DBMS_OUTPUT.PUT_LINE(
                '----------------------------------'
            );

            

        END LOOP;

    END LOOP;

    CLOSE c_facturas;

END;
/
```

### Funcion para cargar N facturas aleatorias con 10*N detalles

```
CREATE OR REPLACE PROCEDURE cargar_datos(
    p_num_facturas IN NUMBER
)
IS
BEGIN

    -- Limpiar datos antiguos
    DELETE FROM detalle;
    DELETE FROM factura;

    COMMIT;

    -- Insertar facturas
    FOR i IN 1 .. p_num_facturas LOOP

        INSERT INTO factura (
            codigo_factura,
            fecha,
            codigo_cliente,
            codigo_vendedor
        )
        VALUES (
            seq_factura.NEXTVAL,
            TRUNC(SYSDATE - DBMS_RANDOM.VALUE(1, 365)),
            TRUNC(DBMS_RANDOM.VALUE(1, 10000)),
            TRUNC(DBMS_RANDOM.VALUE(1, 1000))
        );

    END LOOP;

    COMMIT;

    -- Insertar detalles
    FOR f IN (
        SELECT codigo_factura
        FROM factura
    ) LOOP

        FOR j IN 1 .. 10 LOOP

            INSERT INTO detalle (
                codigo_detalle,
                codigo_factura,
                codigo_producto,
                numero_unidades,
                precio_unitario
            )
            VALUES (
                seq_detalle.NEXTVAL,
                f.codigo_factura,
                TRUNC(DBMS_RANDOM.VALUE(1, 5000)),
                TRUNC(DBMS_RANDOM.VALUE(1, 20)),
                TRUNC(DBMS_RANDOM.VALUE(1000, 50000))
            );

        END LOOP;

    END LOOP;

    COMMIT;

END;
/
```

```
CREATE TABLE benchmark_resultados (
    id            NUMBER GENERATED ALWAYS AS IDENTITY,
    facturas      NUMBER,
    limite        NUMBER,
    tiempo_seg    NUMBER,
    fecha_ejec    DATE DEFAULT SYSDATE
);
```

### Funcion para realizar la pruebas con los distintos parametros
- p_facturas para llenar las tablas con N cantidad de facturas
- p_limit para el LIMIT del bulk collect
- enable_output para determinar si es necesario imprimir, cuando son mucha cantidad de datos ocurre overflow y no se alcanza a ejecutar

```
CREATE OR REPLACE PROCEDURE ejecutar_prueba(
    p_facturas NUMBER,
    p_limit NUMBER
)
IS
    v_inicio NUMBER;
    v_fin NUMBER;
    v_tiempo NUMBER;
BEGIN
    -- Tiempo inicial
    v_inicio := DBMS_UTILITY.GET_TIME;

    -- Ejecutar benchmark
    bulk_collect_facturas(p_limit);

    -- Tiempo final
    v_fin := DBMS_UTILITY.GET_TIME;

    v_tiempo := (v_fin - v_inicio) / 100;

    -- Mostrar resultado
    DBMS_OUTPUT.PUT_LINE(
        'FACTURAS: ' || p_facturas ||
        ' | LIMIT: ' || p_limit ||
        ' | TIEMPO: ' || TO_CHAR(v_tiempo) ||
        ' segundos'
    );

    INSERT INTO benchmark_resultados (
        facturas,
        limite,
        tiempo_seg
    )
    VALUES (
        p_facturas,
        p_limit,
        v_tiempo
    );

    COMMIT;

END;
/
```
